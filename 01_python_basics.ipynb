```python
def calculate_growth(current_sales, previous_sales):
  growth_rate = (current_sales - previous_sales) / previous_sales *100
  return growth_rate

result = calculate_growth(150000, 120000)

print(result)
```

```python
monthly_sales = [12000, 15000, 130000, 180000, 210000]

total_sales = sum(monthly_sales)
average_sales = total_sales / len(monthly_sales)
highest_sales = max(monthly_sales)
lowest_sales = min(monthly_sales)

print("전체 매출:", total_sales)
print("평균 매출:", average_sales)
print("최고 매출:", highest_sales)
print("최저 매출:", lowest_sales)

print("15만 원 이상 매출:")
for sales in monthly_sales:
  if sales >= 150000:
    print(sales)


```
